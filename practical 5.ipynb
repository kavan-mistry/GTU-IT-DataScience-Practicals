### Plot the graph showing the geographical location of students.

```python
import pandas as pd
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('practical data.csv')
```

```python
City_list = list(df['City'].unique())
print(City_list)
```

```python
dd = {}

for i in City_list:
    temp = 0
    for j in list(df['City']):
        if i==j:
            temp = temp + 1
    dd[i] = temp
    
```

```python
city = list(dd.keys())
city_count = list(dd.values())
# city_count = str(city_count)
type(city_count)
```

```python
fig = plt.figure(figsize = (35, 5))

# creating the bar plot
plt.bar(city, city_count, color ='purple', 
        width = 0.2)

plt.xlabel("Cities")

plt.show()
```
