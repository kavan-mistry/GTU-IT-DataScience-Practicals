### Implement decision tree to classify the student as average or clever.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('practical data.csv')
df.head()
```

```python
df['Performance'] = df['PERFORMANCE'].apply(lambda x: True if x=='CLEVER' else False)
df.head()
```

```python
inputs = df[['COA', 'OS', 'CN', 'OOP-1', 'DS', 'EG']]

target = df['Performance']
```

```python
X = inputs
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)
```
