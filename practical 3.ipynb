### Implement a method to find out variation in data. For example the difference between highest and lowest marks in each subject semester wise.

```python
import pandas as pd
import numpy as np
import statistics as stats
```

```python
df = pd.read_csv('practical data.csv')
df.head()
```

```python
# Variance max-min
for i in list(df.columns)[6:]:    
    if(df[i].dtype == 'int'):
        print(i,"---", np.max(df[i]) - np.min(df[i]))
```

```python
for i in list(df.columns)[5:]:
    if(df[i].dtype == 'float64'):
        print(i,"---", np.max(df[i] - np.min(df[i])))
```
