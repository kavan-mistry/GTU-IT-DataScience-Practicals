### Implement a method to treat missing value for gender and missing value for marks.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('practical data.csv')
df
```

```python
print(df.isnull().sum())
```

```python
for i in list(df.columns)[6:12]:
    print(i)
    df[i] = df[i].replace(0, np.nan)
df
```

```python
for i in list(df.columns)[6:12]:
    df[i].fillna(df[i].mean(), inplace=True)
```
