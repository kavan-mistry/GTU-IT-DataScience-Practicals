### Implement Logistic regression to classify the student as average or clever.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
df = pd.read_csv("practical data.csv")
df.head()
```

```python
df['Performance'] = df['PERFORMANCE'].apply(lambda x: True if (x == 'CLEVER') else False) 
df.head()
```

```python
X = df[['COA', 'OS', 'CN', 'OOP-1', 'DS', 'EG']]
y = df.Performance

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

model = LogisticRegression()

model.fit(X_train, y_train)
```
