### Plot the graph showing number of male and female students.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('practical data.csv')
df.head()
```

```python
male = 0
female = 0

for i in df['Gender']:
    if i == 'Male':
        male = male + 1
    if i == 'Female':
        female = female + 1
        
km = {}
km['male'] = male
km['female'] = female
km
```

```python
fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(list(km.keys()), list(km.values()), color ='purple',
        width = 0.3)

plt.xlabel("Courses offered")
plt.ylabel("Male/Female")
plt.show()
```
